```python
from non_dominated_sorting import fast_non_dominated_sort
import os.path
import numpy as np
import matplotlib.pyplot as plt
from hv import Hypervolume
from normalization import normalize
import pandas as pd

# the result folder as a path
folder = os.path.join(os.path.dirname(os.path.dirname(os.getcwd())), "ttp-morl/evaluations/f_files")

# all submissions received
participants = [   
                      "PN-PHN",
                      "AM-PHN",
                    "AM-MBPS",
                    "PN-MBPS"
               ]
```

```python
nondom_f_list = [None for i in range(1080)]
all_f_dict = {}
for participant in participants:
    participant_path = os.path.join(folder, participant)
    num_files = len(os.listdir(participant_path))
    all_f_dict[participant] = [None for i in range(num_files)]
#     print(os.listdir(participant_path))
    for i in range(-1, num_files-1):
        filename = participant+"_"+str(i)+".pt.npy"
        file_path = os.path.join(participant_path, filename)
        f_list = np.load(file_path)
        all_f_dict[participant][i]= f_list
        for j in range(1080):
            if nondom_f_list[j] is None:
                nondom_f_list[j] = f_list[:,j,:]
                continue
            nondom_f_list[j] = np.concatenate([nondom_f_list[j], f_list[:,j,:]], axis=0)
            nondom_idx = fast_non_dominated_sort(nondom_f_list[j])[0]
            nondom_f_list[j] = nondom_f_list[j][nondom_idx, :]
```

```python
utopia_list = [None for i in range(1080)]
nadir_list = [None for i in range(1080)]
for i in range(1080):
    utopia_list[i] = np.min(nondom_f_list[i], axis=0, keepdims=True)
    nadir_list[i] = np.max(nondom_f_list[i], axis=0, keepdims=True)
#     print(utopia_list[i], nadir_list[i], utopia_list[i] == nadir_list[i])
```

```python
all_score_dict = {}
for participant in participants:
    participant_path = os.path.join(folder, participant)
    num_epochs = len(all_f_dict[participant])
    print(participant, num_epochs)
    all_score_dict[participant] = np.zeros((num_epochs, 1080))
#     print(os.listdir(participant_path))
    for i in range(-1, num_epochs-1):
        f_list = all_f_dict[participant][i]
        for j in range(1080):
            f = f_list[:,j,:]
            u, n = utopia_list[j], nadir_list[j]
            n[n==u] = n[n==u]+10
            diff = n-u
#             diff[diff==0] = 1
            f_norm = (f-u)/diff
            score = Hypervolume(np.array([1,1])).calc(f_norm)
            all_score_dict[participant][i,j]=score
    plt.plot(all_score_dict[participant].mean(axis=1), label=participant)
plt.xlabel("Epoch")
plt.ylabel("HV")
plt.legend()
plt.gcf()
plt.savefig("mo-training.jpg", dpi=200, bbox_inches='tight', )
plt.show()
```
