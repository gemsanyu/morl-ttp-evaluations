```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tensorflow.python.summary.summary_iterator import summary_iterator
# import seaborn as sns
from scipy import stats
import tensorboard as tb
```

```python
file_path_list = [
    "runs/PN/events.out.tfevents.1683857930.komputasi07.27856.0",
    "runs/AM/events.out.tfevents.1683721923.komputasi04.14911.0"
]
model_list = ["PN","AM"]
c_dict = {"PN":"blue", "AM":"red"}
duration_list = []
df_dict = {}
for i,file_path in enumerate(file_path_list):
    model = model_list[i]
    value_dict = {}
    wt_list = []
    for summary in summary_iterator(file_path):
        value = summary.summary.value
        if len(value)==0:
            continue
        tag = value[0].tag
        val = value[0].simple_value
        if not tag in value_dict.keys():
            value_dict[tag] = []
        value_dict[tag] += [val]
        wt_list += [summary.wall_time]
    wt_list = np.asanyarray(wt_list)
    duration_list += [np.max(wt_list) - np.min(wt_list)]
#     print(model_list[i], np.min(wt_list), np.max(wt_list))
    df = pd.DataFrame.from_dict(value_dict)
    df_dict[model] = df
```

```python
print(duration_list)
for i, duration in enumerate(duration_list):
    days = int(duration / (24*3600))
    hours = int((duration % (24*3600))/3600)
    minutes = int((duration % (3600))/60)
    print(model_list[i], days, hours, minutes)
```

```python
ax = plt.subplot()
for model, df in df_dict.items():
    epochs = np.arange(len(df["Validation Total Cost"]))+1
#     print(model, epochs)
    ax.plot(epochs, df["Validation Total Cost"].ewm(alpha=0.1).mean(), label=model, c=c_dict[model], linewidth=1)
    ax.plot(epochs, df["Validation Total Cost"], alpha=0.1, c=c_dict[model])
ax.set_xlabel("Epoch")
ax.set_xlim(left=0)
ax.set_ylabel("Total cost")
ax.legend()
plt.gcf()
plt.savefig("training-so-total-cost.jpg", dpi=200, bbox_inches='tight', )
plt.show()
```

```python
#now training progress DRLMOA
import os
import datetime

model = "AM-DRLMOA"
dict_list = []
for i in range(1,101):
    dir_i = "runs/"+model+"/"+model+str(i)+"_100"
    epoch_dict = {}
    for runs_file in os.listdir(dir_i):
        file_path = dir_i+"/"+runs_file
        for summary in summary_iterator(file_path):
            value = summary.summary.value
            if len(value)==0:
                continue
            if summary.wall_time < 1683905310:
                continue
            if not summary.step in epoch_dict.keys():
                epoch_dict[summary.step] = {"model":"AM-DRLMOA", "weight":i, "walltime":summary.wall_time}
            epoch_dict[summary.step]["epoch"] = int(summary.step)
            tag = value[0].tag
            val = value[0].simple_value
            epoch_dict[summary.step][tag] = val
    for epoch in epoch_dict.keys():
        epoch_dict[epoch]["weight"] = i
        dict_list += [epoch_dict[epoch]]
df = pd.DataFrame(dict_list)
```

```python
import os
import datetime     
model = "PN-DRLMOA"
directory = "runs/"+model+"/"+model
dict_list = []
epoch_dict = {}
weight = 1
last_epoch = -1
new_dict = None
for runs_file in os.listdir(directory):
    file_path = directory+"/"+runs_file
    for summary in summary_iterator(file_path):
        value = summary.summary.value
        if len(value)==0:
            continue
        if summary.wall_time < 1683905310:
            continue
        current_epoch = summary.step
        if current_epoch != last_epoch:
            if new_dict is not None:
                dict_list += [new_dict]
            if current_epoch < last_epoch:
                weight+=1
            last_epoch = current_epoch
            new_dict = {"model":"PN-DRLMOA", "weight":weight, "epoch":current_epoch, "walltime":summary.wall_time}
        tag = value[0].tag
        val = value[0].simple_value
        new_dict[tag] = val
dict_list += [new_dict]
df_pn = pd.DataFrame(dict_list)
# print(df_pn)
```
